# Activation Functions in Neural Networks

Activation functions add essential non-linearity to neural networks. Without them, even deep networks would merely be linear models. Here are the key activation functions:

### 1. Sigmoid (Logistic)
- Mathematical form: `f(x) = 1 / (1 + e^(-x))`
- Output range: (0, 1)
- Characteristics: Smooth, bounded function commonly used in binary classification

### 2. TanH (Hyperbolic Tangent)
- Mathematical form: `f(x) = (e^x - e^(-x)) / (e^x + e^(-x))`
- Output range: (-1, 1)
- Characteristics: Similar to sigmoid but centered around 0, useful when negative outputs are needed

### 3. ReLU (Rectified Linear Unit)
- Mathematical form: `f(x) = max(0, x)`
- Output range: [0, ∞)
- Characteristics: 
  - Simple computation
  - No vanishing gradient for positive inputs
  - Zero gradient for negative inputs
  - Most commonly used in modern networks

### 4. Leaky ReLU
- Mathematical form: `f(x) = max(kx, x)` where k is a small constant
- Output range: (-∞, ∞)
- Characteristics:
  - Addresses "dying ReLU" problem
  - Allows small negative gradients
  - Parameter k determines the slope for negative inputs

````python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, k=0.1):
    return np.where(x > 0, x, k * x)

# Create input values
x = np.linspace(-5, 5, 200)

# Create subplot grid
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Neural Network Activation Functions', fontsize=16)

# Plot Sigmoid
ax1.plot(x, sigmoid(x), 'b-', label='Sigmoid')
ax1.grid(True)
ax1.set_title('Sigmoid')
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.legend()
ax1.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax1.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Plot TanH
ax2.plot(x, tanh(x), 'r-', label='TanH')
ax2.grid(True)
ax2.set_title('TanH')
ax2.set_xlabel('x')
ax2.set_ylabel('f(x)')
ax2.legend()
ax2.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax2.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Plot ReLU
ax3.plot(x, relu(x), 'g-', label='ReLU')
ax3.grid(True)
ax3.set_title('ReLU')
ax3.set_xlabel('x')
ax3.set_ylabel('f(x)')
ax3.legend()
ax3.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax3.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Plot Leaky ReLU
ax4.plot(x, leaky_relu(x), 'm-', label='Leaky ReLU')
ax4.grid(True)
ax4.set_title('Leaky ReLU')
ax4.set_xlabel('x')
ax4.set_ylabel('f(x)')
ax4.legend()
ax4.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax4.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Adjust layout and display
plt.tight_layout()
plt.show()
````

# Neural Network Layers: A Comprehensive Overview

## 1. Fully Connected (Linear) Layer
- Every neuron in the current layer connects to every neuron in the subsequent layer
- Basic building block of traditional neural networks
- Key characteristics:
  - Dense connectivity pattern
  - Suitable for processing flat, non-structured data
  - High parameter count due to full connectivity

![Neural Network Architecture](https://miro.medium.com/v2/resize:fit:720/format:webp/1*VHOUViL8dHGfvxCsswPv-Q.png)

## 2. Convolutional Layer (Conv)
- Applies sliding filters/kernels across input data
- Primary component of CNNs (Convolutional Neural Networks)
- Key characteristics:
  - Local connectivity pattern
  - Parameter sharing across spatial dimensions
  - Particularly effective for image processing
  - Maintains spatial relationships in data

![Deep Learning Framework](https://miro.medium.com/v2/resize:fit:720/format:webp/0*HRiuP0ZsyHkRR5AL.png)

## 3. Recurrent Layer
- Contains loops/cycles in connections
- Designed for sequential data processing
- Key characteristics:
  - Memory capability through recurrent connections
  - Can process variable-length sequences
  - Maintains internal state
  - Useful for time series, text, and sequential data

![Deep Learning Structure](https://miro.medium.com/v2/resize:fit:640/format:webp/0*VGyW0ISyt0VESxDC.png)

## 4. DeConvolutional Layer (Transposed Convolution)
- Reverses the spatial transformation of convolution
- Used in generative models and upsampling
- Key characteristics:
  - Increases spatial dimensions
  - Important for image generation/reconstruction
  - Common in autoencoders and GANs

![DeConvolution Structure](https://miro.medium.com/v2/resize:fit:640/format:webp/1*8MhRh4T970Ewp4EYHq4aEQ.gif)

## 5. Pooling Layer
- Reduces spatial dimensions through downsampling
- Common variants: Max, Min, and Average pooling
- Key characteristics:
  - No learnable parameters
  - Reduces computation and prevents overfitting
  - Provides spatial invariance

The 16 extra samples result from how PyTorch handles the division of the dataset into batches when the total number of samples is not perfectly divisible by the batch size.

In the example:

### Key Points:
1. **Dataset Size**: The training dataset has `6000` samples.
2. **Batch Size**: Each batch contains `32` samples.
3. **Number of Batches**:
   - Dividing the dataset size by the batch size: $ \text{6000} \div \text{32} = 187.5 $.
   - Since the number of batches must be an integer, PyTorch rounds up to the next integer (`188`) to include all samples. This is known as the **last incomplete batch**.
   - The last batch will have fewer samples than the batch size.

4. **Total Samples in DataLoader**:
   - For `len(train_loader) = 188`, multiplying by the batch size gives $ \text{188} \times \text{32} = 6016$, which is 16 more than the actual dataset size (`6000`).

5. **Explanation of Extra Samples**:
   - By default, **`shuffle=True`** ensures that the dataset is shuffled, and PyTorch will include samples to fill the last batch. To handle this, it samples randomly from the dataset to fill the shortfall (repeating some data).

### How to Fix It:
To avoid including extra samples in the last batch, you can set `drop_last=True` in the `DataLoader`. This ensures that any incomplete batch is dropped.

```python
train_loader = DataLoader(dataset=train_dataset, batch_size=32, shuffle=True, drop_last=True)
```

### Result with `drop_last=True`:
1. `len(train_loader)` will be $\lfloor \text{6000} \div \text{32} \rfloor = 187 $.
2. Total samples processed: $ 187 \times 32 = 5984 $, and the 16 remaining samples are excluded.

If including all samples is important, keep `drop_last=False` (default), but note that the last batch may be smaller than the batch size.

### Training a neural network using PyTorch

We will train a multiclass classifier on the MNIST dataset in this section.

````python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset, random_split
from torchvision import datasets
from torchsummary import summary

import numpy as np
import matplotlib.pyplot as plt
````

````python
# A class in Python is a blueprint for creating objects. It defines a set of attributes (data) and methods (functions)
# that an object of the class will have. Classes allows to implement object-oriented programming (OOP) 
# by encapsulating data and behavior into a single structure.

class ConvNet(nn.Module): # nn.Module is the base class for all neural network modules in PyTorch
    def __init__(self): # The __init__ method is a special method in Python classes that is called when a new object of the class is created
        super(ConvNet, self).__init__() # The super() function is used to call the __init__() method of the parent class
        self.conv_1 = nn.Conv2d(1, 16,  3, 1) # 1 channel input, 16 channels output, 3x3 kernel, stride 1
        self.conv_2 = nn.Conv2d(16, 32, 3, 1) # 16 channels input, 32 channels output, 3x3 kernel, stride 1
        self.dp1 = nn.Dropout2d(0.10)
        self.dp2 = nn.Dropout(0.25)
        self.fc1 = nn.Linear(4608, 64) # 4608 is the number of features after flattening the tensor 12*12*32
        self.fc2 = nn.Linear(64, 10)

    def forward(self, x): # The forward method defines the forward pass of the neural network
        x = self.conv_1(x)
        x = F.relu(x) # Applies the rectified linear unit function element-wise
        x = self.conv_2(x)
        x = F.relu(x)
        x = F.max_pool2d(x, 2) # Applies a 2D max pooling over an input signal composed of several input planes
        x = self.dp1(x)
        x = torch.flatten(x, 1)
        x = self.fc1(x)
        x = F.relu(x)
        x = self.dp2(x)
        x = self.fc2(x)
        output = F.log_softmax(x, dim=1) # Applies a softmax followed by a logarithm function, that is F.log_softmax(x) = log(exp(x) / exp(x).sum())

        return output
````

````python
from torchviz import make_dot

# Create a sample input
sample_input = torch.randn(1, 1, 28, 28)  # Batch size=1, Channels=1, Height=28, Width=28

# Instantiate the model
model = ConvNet()
# Forward pass
output = model(sample_input)
# Visualize
make_dot(output, params=dict(model.named_parameters())).render("convnet", format="png")
````

### 1. **Input Dimensions**:
- The input to the `ConvNet` is an image tensor with dimensions:
  $$
  \text{Batch size} \times \text{Channels} \times \text{Height} \times \text{Width}
  $$
  In your case, the input size is assumed to be:
  $$
  1 \times 1 \times 28 \times 28
  $$
  - **1** channel (grayscale image).
  - **28 × 28**: Height and width of the image.

### 2. **Conv Layer 1** (`self.conv_1`):
```python
self.conv_1 = nn.Conv2d(1, 16, 3, 1)
```
- **Parameters**:
  - Kernel size: \(3 \times 3\)
  - Stride: \(1\)
  - Padding: \(0\) (default, no padding).
- **Formula for output size**:
  $$
  \text{Output Size} = \frac{\text{Input Size} - \text{Kernel Size}}{\text{Stride}} + 1
  $$
  For \(28 \times 28\):
  $$
  \text{Height (or Width)} = \frac{28 - 3}{1} + 1 = 26
  $$
  The output after `conv_1`:
  $$
  16 \times 26 \times 26 \quad (\text{16 channels, 26×26 spatial dimensions})
  $$

### 3. **Conv Layer 2** (`self.conv_2`):
```python
self.conv_2 = nn.Conv2d(16, 32, 3, 1)
```
- **Parameters**:
  - Kernel size: \(3 \times 3\)
  - Stride: \(1\)
  - Padding: \(0\).
- **Input** to this layer is \(16 \times 26 \times 26\).
- Using the same formula:
  $$
  \text{Height (or Width)} = \frac{26 - 3}{1} + 1 = 24
  $$
  The output after `conv_2`:
  $$
  32 \times 24 \times 24 \quad (\text{32 channels, 24×24 spatial dimensions})
  $$

### 4. **Max Pooling** (`F.max_pool2d(x, 2)`):
- A \(2 \times 2\) max-pooling operation with default stride \(2\) reduces the height and width by half.
- Input to the max-pooling is \(32 \times 24 \times 24\).
- **Formula for max-pooling output**:
  $$
  \text{Output Size} = \frac{\text{Input Size}}{\text{Pooling Kernel Size}}
  $$
  For \(24 \times 24\):
  $$
  \text{Height (or Width)} = \frac{24}{2} = 12
  $$
  The output after max pooling:
  $$
  32 \times 12 \times 12
  $$

### 5. **Flattening**:
The output of the max-pooling layer is \(32 \times 12 \times 12\). To pass it to the fully connected layer (`self.fc1`), the tensor is **flattened**:
$$
32 \times 12 \times 12 = 4608 \quad (\text{features passed to `fc1`}).
$$

### Summary of Shapes:
| **Layer**            | **Output Shape**          | **Explanation**                     |
|-----------------------|---------------------------|-------------------------------------|
| Input Image           | $1 \times 28 \times 28$  | Input size (grayscale image).       |
| After `conv_1`        | $16 \times 26 \times 26$ | 3x3 kernel, no padding, stride 1.   |
| After `conv_2`        | $32 \times 24 \times 24$ | 3x3 kernel, no padding, stride 1.   |
| After Max Pooling     | $32 \times 12 \times 12$ | 2x2 pooling, stride 2.              |
| Flattened Tensor      | $4608$                   | Flattened for `fc1`.               |

### Kernel Size in Convolutional Layers

- **Kernel Size of 3x3**:
  - Kernel sizes are typically odd to ensure the input image pixels are symmetrically distributed around a central pixel.
  - A 1x1 kernel would only consider a single pixel, which misses information from neighboring pixels.
  - A kernel size of 3x3 is a good balance, allowing the kernel to capture local visual patterns while being small enough for efficient computation.
  
- **Why Not Larger Kernels (e.g., 5x5, 7x7, 27x27)?**:
  - Larger kernels (like 27x27) may result in very coarse-grained features, which are not ideal for capturing fine-grained visual patterns.
  - The most important visual features are often local, so a smaller kernel (like 3x3) is preferred to capture these details.

- **Consecutive 3x3 Convolutional Layers**:
  - Using two 3x3 convolutional layers is functionally equivalent to using a single 5x5 kernel, but it has advantages:
    - **Deeper Networks**: More layers can capture more complex features.
    - **Fewer Parameters**: Smaller kernels result in fewer parameters to train.
    - **Specialized Kernels**: Multiple small kernels may specialize in detecting different visual patterns (e.g., edges, circles, colors).

- **Number of Channels**:
  - The output of a convolutional layer often has more channels than the input. For example, the first convolutional layer outputs 16 channels, each detecting different types of features.
  - The second layer increases the number of channels to 32, aiming to capture more complex patterns.

- **Stride of 1**:
  - A stride of 1 is ideal when using a small kernel (e.g., 3x3) because it ensures the kernel scans all pixels in the image without skipping any.
  - Larger strides (e.g., 10) would skip many pixels, which can lead to a loss of detail. Larger stride values may be used for larger kernel sizes (e.g., 100x100).

````python
# The model could have been written using the following torch.nn.Sequential API:

model_1 = nn.Sequential(
    nn.Conv2d(1, 16, 3, 1),
    nn.ReLU(),
    nn.Conv2d(16, 32, 3, 1),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Dropout2d(0.10),
    nn.Flatten(),
    nn.Linear(4608, 64),
    nn.ReLU(),
    nn.Dropout(0.25),
    nn.Linear(64, 10),
    nn.LogSoftmax(dim=1)
)
````

### Define training and inference routines

````python
# Define the training loop

def train_model(model, device, train_dataloader, optim, epoch):
    model.train() # Set the model to training mode
    for batch_idx, (data, target) in enumerate(train_dataloader):
        data, target = data.to(device), target.to(device) # Move the data and target tensors to the device (GPU)
        optim.zero_grad() # Zero the gradients
        pred_prob = model(data) # Forward pass
        loss = F.nll_loss(pred_prob, target) # Calculate the negative log likelihood loss: -log(probability)
        loss.backward() # Backward pass which computes the gradients
        optim.step() # Update the weights
        if batch_idx % 100 == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_dataloader.dataset),
                100. * batch_idx / len(train_dataloader), loss.item()))
            
# Define the test loop

def test_model(model, device, test_dataloader):
    model.eval() # Set the model to evaluation mode
    test_loss = 0
    success = 0
    with torch.no_grad(): # Disable gradient tracking
        for data, target in test_dataloader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            success += pred.eq(target.view_as(pred)).sum().item() # Check if the prediction is correct

    test_loss /= len(test_dataloader.dataset) # Calculate the average loss
    accuracy = 100. * success / len(test_dataloader.dataset) # Calculate the accuracy
    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, success, len(test_dataloader.dataset), accuracy))
````

### Summary of the Explanation on Batch Size and Dataset Normalization

- **Batch Size**:
  - **Common Choice**: A batch size of **32** is a commonly used value for many deep learning models.
  - **Trade-off**: 
    - **Small Batch Size**: Leads to frequent gradient updates but can result in noisy gradients, making the learning process slow.
    - **Large Batch Size**: May increase waiting time for gradient calculations, slowing down the training process, as a single large batch may take a long time to process.
  - **Recommendation**: Frequent updates with smaller batches typically lead to better optimization, as they balance the trade-off between noisy updates and waiting time.

- **Dataset Configuration**:
  - **Local Storage**: The dataset is stored locally, and a specific batch size is set for both training and testing datasets.
  - **Shuffling**: For the **training dataset**, random shuffling ensures that the data is evenly distributed across batches, avoiding any order bias.
  - **Normalization**: 
    - **Normalization** helps in transforming the dataset to have a **mean of 0** and a **standard deviation of 1**.
    - **Training from Scratch**: If training from scratch, the mean and standard deviation are computed from the training data.
    - **Transfer Learning**: If using a pre-trained model for transfer learning, the normalization values (mean and standard deviation) are typically taken from the pre-trained model’s original dataset (e.g., ImageNet for image-based models).

This approach balances training efficiency and model accuracy, ensuring the model learns effectively from the data while avoiding issues like noisy gradients or long training times due to excessively large batch sizes.

### Optimization schedule

1. **Stochastic Gradient Descent (SGD) and Its Variants**

    - **Stochastic Gradient Descent (SGD)**:
      - Updates the model parameters in the following manner:
        $$
        \mathbf{B} = \mathbf{B} - \alpha \frac{\partial L(\mathbf{X}, \mathbf{y})}{\partial \mathbf{B}}
        $$
        - **Where**:
          - $\mathbf{B}$: Parameters of the model
          - $\mathbf{X}$: Input training data
          - $\mathbf{y}$: Corresponding labels
          - $L$: Loss function
          - $\alpha$: Learning rate
        - **Operation**: SGD performs this update for every training example pair $(\mathbf{X}, \mathbf{y})$.
    
    - **Mini-batch Gradient Descent**:
      - A variant of SGD, which performs updates for every **k** examples (batch size).
      - Gradients are calculated collectively for the whole mini-batch instead of individual training examples.
    
    - **Batch Gradient Descent**:
      - Another variant where parameter updates are performed by calculating the gradient across the entire dataset.

2. **Adagrad: Adaptive Learning Rates for Parameters**

  In the previous optimization schedule, a **single learning rate** was used for all the parameters of the model. However, in cases where different parameters need to be updated at different paces (especially with sparse data), some parameters may be more involved in feature extraction than others. **Adagrad** introduces the concept of **per-parameter updates**.

  The update rule in Adagrad is given by:

  $$
  \mathbf{B}_i^{t+1} = \mathbf{B}_i^t - \frac{\alpha}{\sqrt{G_{i}^t + \epsilon}} \cdot \frac{\partial L}{\partial \mathbf{B}_i}
  $$

  Where:

  - $\mathbf{B}_i$: The $i$-th parameter of the model.
  - $G_i^t$: The sum of squared gradients for the $i$-th parameter from time step $0$ to $t$, i.e., 
    $$
    G_i^t = \sum_{\tau=1}^t \left(\frac{\partial L}{\partial \mathbf{B}_i}\right)^2
    $$
  - $\alpha$: The global learning rate.
  - $\epsilon$: A small value added for numerical stability (to avoid division by zero).
  - $\frac{\partial L}{\partial \mathbf{B}_i}$: The gradient of the loss function with respect to the $i$-th parameter.

  - How Adagrad Works:
    - **Per-parameter update**: The learning rate for each parameter is adjusted based on the historical gradients. Parameters with frequent updates (large gradients) will have smaller learning rates, while those with infrequent updates (small gradients) will have larger learning rates.
    - **Benefits**: This approach ensures that frequently changing parameters update less aggressively, while less active parameters are updated more rapidly. This is especially useful in models trained on sparse data where only a few features are frequently used.

3. **Adadelta: Resolving the Learning Rate Decay Problem**

  In **Adagrad**, the learning rate's denominator continuously increases due to the accumulation of squared gradients at each time step. This leads to the problem of the learning rate decaying to very small values. **Adadelta** addresses this issue by limiting the accumulated sum of squared gradients to a fixed window of previous time steps. This is done by computing a **running decaying average** of past gradients.

  - Adadelta Update Rule:

    The sum of squared gradients is computed as:

  $$
  G_i^{t} = \gamma \cdot G_i^{t-1} + (1 - \gamma) \cdot \left( \frac{\partial L}{\partial \mathbf{B}_i} \right)^2
  $$

Where:

- $ G_i^{t}$: The sum of squared gradients for the $i$-th parameter at time step $t$.
- $\gamma$: The decaying factor (typically close to 1) that controls the moving average of previous squared gradients.
- $ \frac{\partial L}{\partial \mathbf{B}_i}$: The gradient of the loss function with respect to the $i$-th parameter.

With this formulation, we ensure that the sum of squared gradients doesn't grow indefinitely, thanks to the decaying average.

  - Parameter Update Equation:

The parameter update is given by:

$$
\Delta \mathbf{B}_i = \mathbf{B}_i^{t} - \mathbf{B}_i^{t-1} = - \frac{\alpha \cdot \frac{\partial L}{\partial \mathbf{B}_i}}{\sqrt{G_i^t + \epsilon}}
$$

Where:

-$\Delta \mathbf{B}_i$: The change in the \(i\)-th model parameter.
-$G_i^t$: The sum of squared gradients for the \(i\)-th parameter.

- Adjusting for Dimensionality:

One issue in **Adagrad** (and other methods) is that the **root mean squared gradient** isn't dimensionless, and using it directly as a coefficient for the learning rate could lead to problems. To handle this, **Adadelta** introduces a second running average, this time for the **squared parameter updates**.

The squared parameter updates are computed as:

$$
U_i^{t} = \gamma \cdot U_i^{t-1} + (1 - \gamma) \cdot \left( \Delta \mathbf{B}_i \right)^2
$$

Where:

- $ U_i^t$: The sum of squared parameter updates for the \(i\)-th parameter at time step \(t\).
- $\gamma$: The decaying factor, just like in the gradient accumulation.

- Final Adadelta Update Equation:

The final Adadelta update rule is:

$$
\mathbf{B}_i^{t+1} = \mathbf{B}_i^t - \frac{\sqrt{U_i^t + \epsilon}}{\sqrt{G_i^t + \epsilon}} \cdot \frac{\partial L}{\partial \mathbf{B}_i}
$$

Where:

-$\frac{\sqrt{U_i^t + \epsilon}}{\sqrt{G_i^t + \epsilon}}$: This term normalizes the updates to account for both the sum of squared gradients and the sum of squared parameter updates.

- How Adadelta Works:

  - **Key Improvement**: By maintaining a decaying average of squared gradients and parameter updates, **Adadelta** avoids the problem of continuously decaying learning rates seen in **Adagrad**.
  - **No Learning Rate Tuning**: Unlike **Adagrad**, **Adadelta** eliminates the need to manually tune the learning rate $\alpha$, as it is internally adjusted based on the moving averages of gradients and parameter updates.
  - **Adaptive Updates**: Like **Adagrad**, **Adadelta** adapts the learning rates for each parameter, but in a more controlled way using the running averages. This ensures that the learning rate does not diminish excessively over time, making **Adadelta** more stable in the long run.
  - As mentioned earlier, the final **Adadelta** equation does not require an explicit learning rate. However, you can still provide a learning rate as a multiplier, but it is not mandatory. The only key hyperparameter for **Adadelta** is the decaying factor, which determines how much the previous gradients and parameter updates influence the current update.

4. RMSprop:

  - **RMSprop** is closely related to **Adadelta**, with one key distinction: **RMSprop** does not correct for the dimensionality problem (i.e., the scaling issue of the root mean squared gradient). As a result, the update equation in **RMSprop** remains the same as **Adagrad's** (Equation 1.6), where the sum of squared gradients (\(SSG\)) is computed as described in **Adadelta** (Equation 1.7).
  
  - **Key Differences from Adadelta**:
    - **RMSprop** requires both a **base learning rate** and a **decaying factor**, whereas **Adadelta** does not explicitly require a learning rate.

5. Adaptive Moment Estimation (Adam):

**Adam** is another optimization algorithm that computes **customized learning rates** for each parameter. It combines the ideas from **Adagrad** and **RMSprop** by using the **decaying average of previous squared gradients** (like **Adadelta** and **RMSprop**), but it also introduces the **decaying average of previous gradient values**, making it an extension of **RMSprop**.

  - The update rule for **Adam** is:

  $$
  \mathbf{G}_i^{t} = \beta_1 \cdot \mathbf{G}_i^{t-1} + (1 - \beta_1) \cdot \frac{\partial L}{\partial \mathbf{B}_i}
  $$

  $$
  \mathbf{V}_i^{t} = \beta_2 \cdot \mathbf{V}_i^{t-1} + (1 - \beta_2) \cdot \left(\frac{\partial L}{\partial \mathbf{B}_i}\right)^2
  $$

  Where:

  - $\mathbf{G}_i^{t}$: The decaying average of gradients.
  - $\mathbf{V}_i^{t}$: The decaying average of squared gradients.
  - $\beta_1$: The decaying factor for the gradients.
  - $\beta_2$: The decaying factor for the squared gradients.

  These terms are similar to the **first** and **second moments** of the gradient, which is why this method is called **adaptive moment estimation (Adam)**.

- Bias Correction:

Since the initial values of both the gradient and squared gradient averages are biased toward zero (especially in the early stages of training), **bias correction** is applied to counteract this issue:

- For the gradient term:

  $$
  \hat{\mathbf{G}}_i^{t} = \frac{\mathbf{G}_i^{t}}{1 - \beta_1^t}
  $$

- For the squared gradient term:

  $$
  \hat{\mathbf{V}}_i^{t} = \frac{\mathbf{V}_i^{t}}{1 - \beta_2^t}
  $$

- Adam Parameter Update:

Once the bias-corrected values are calculated, the parameter update rule is given by:

$$
\mathbf{B}_i^{t+1} = \mathbf{B}_i^t - \frac{\alpha \cdot \hat{\mathbf{G}}_i^{t}}{\sqrt{\hat{\mathbf{V}}_i^{t}} + \epsilon}
$$

Where:

- $\hat{\mathbf{G}}_i^{t}$ is the bias-corrected first moment estimate (the decaying average of the gradient).
- $\hat{\mathbf{V}}_i^{t}$ is the bias-corrected second moment estimate (the decaying average of the squared gradient).
- $\alpha$ is the **base learning rate**.
- $\epsilon$ is a small value to avoid division by zero (typically $10^{-8}$.

- Hyperparameters:

  - **Base Learning Rate** $\alpha$: Controls the size of each update.
  - **Decaying Factor for Gradient** $\beta_1$: Typically close to 1 (e.g., 0.9).
  - **Decaying Factor for Squared Gradient** $\beta_2$: Typically close to 1 (e.g., 0.999).

- Advantages of Adam:

  - **Self-adjusting learning rates** for each parameter.
  - **Bias correction** helps stabilize the early phases of training.
  - Generally, **Adam** is one of the most successful and widely used optimization algorithms in deep learning, particularly for training complex models.

The choice of optimizer depends on your dataset and model. For sparse data, adaptive optimizers like **Adagrad**, **RMSprop**, **Adadelta**, and **Adam** are preferred as they adjust the learning rate per parameter. **SGD** can also work but may take longer to converge. **Adam** is often the best choice due to its combination of first and second moment calculations and bias correction, making it robust for complex models. Experimenting with different optimizers can help you find the best balance between training time and model performance.
